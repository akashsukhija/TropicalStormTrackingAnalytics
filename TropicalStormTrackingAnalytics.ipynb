```python
#Clean Code
def categories(Max_wind,cat_storm,year):
    if Max_wind >= 64:
        cat_storm[year][1] += 1
    if Max_wind >= 83:
        cat_storm[year][2] += 1
    if Max_wind >= 96:
        cat_storm[year][3] += 1
    if Max_wind >= 113 :
        cat_storm[year][4] += 1
    if Max_wind >= 137:
        cat_storm[year][5] += 1
        
    return cat_storm

def Max_sustained_wind(storm_dict,lst2):
    for key,value in storm_dict.items():
        lst2.append(value[6])
    Max_wind=int(max(lst2))
    print('Maximum Sustained wind in Knots:', Max_wind)
    for key,value in storm_dict.items():
        if int(value[6]) == Max_wind:
            print('Date:',value[0])
            print('Time:',value[1])
            break
    return Max_wind
    
def Pressure_change(storm_dict,lst4):
    for key,value in storm_dict.items():
        if value[7].strip() != '-999':
            lst4.append(value[7])
        if lst4 == []:
            change = ('Missing Values')
        else:
            change = int(max(lst4)) - int(min(lst4))
    return change

def Landfall_count(storm_dict):
    landfall_count=[]
    for key,value in storm_dict.items():
        if value[2].strip() == 'L':
            landfall_count.append(value[2])
    count=len(landfall_count)
    return count



x = "hurdat2-1851-2018-120319.txt"
y = "hurdat2-nepac-1949-2018-122019.txt"
m = [x,y]   #Storm Data Files

for p in range (0,2):
    with open(m[p]) as file :

        n = 0
        line_number = 0
        list2 = []
        a = 0
        cat_storm = {}
        year=0
        for i in range(1851,2019):
            cat_storm[i] = [0, 0, 0, 0, 0 ,0]
        for line in file:
            line = line.strip()
            if line.startswith("AL") or line.startswith("EP"):
            
                storm_dict = {}
                list1 = line.split(",")
                year=int(list1[0][4:8])
                print('Storm ID:', list1[0])
                if year in cat_storm.keys():
                    cat_storm[year][0]+=1
                
                lst2 = []
                lst4 = []
                change = 0
                if list1[1].strip() != 'UNNAMED':
                    print('Storm Name:', list1[1].strip())
                n = int(list1[2])
                line_number = 0
            else:
                #print(line)
                storm_dict[line_number] = line.split(",") #Assigning Line Number as Key and Splitted Line as Values
                line_number += 1
                if line_number == n:
                    print ('Start Date:', storm_dict[0][0])
                    print('End Date:', storm_dict[n-1][0])
                
                    Max_wind=Max_sustained_wind(storm_dict,lst2)
                    cat_storm=categories(Max_wind,cat_storm,year)
                
                #Once The List is checked for a particular storm id it should find the category as well increase counter to the storm_count
                    change=Pressure_change(storm_dict,lst4)
                    print('The Total Pressure Change in Millibars:', change)
                
                    count=Landfall_count(storm_dict)
                    print(('Total Number of Times it Had Landfall:'), count)
                    print('---------------------------------------------')
    

    print('Year Storms Cat.1 Cat.2 Cat.3 Cat.4 Cat.5')
    for key, value in cat_storm.items():
        print('{} {} {} {} {} {} {}'.format(key, value[0], value[1], value[2], value[3], value[4], value[5]))
```
